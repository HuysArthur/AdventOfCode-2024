```python
sample_input = """
....#.....
.........#
..........
..#.......
.......#..
..........
.#..^.....
........#.
#.........
......#..."""
```

```python
with open("input.txt") as f:
    real_input =(f.read())
```

```python
def read_area(input: str) -> list[list[str]]:
    array = []

    rows = input.strip().split('\n')
    for row in rows:
        array.append(list(row))

    return array
```

```python
UP = 0
RIGHT = 1
DOWN = 2
LEFT = 3
```

```python
def out_of_area(position, area):
    height = len(area)
    widht = len(area[0])

    return not (0 <= position[0] < widht and 0 <= position[1] < height)

def on_obstacle(position, area):
    return area[position[1]][position[0]] == '#'
```

```python
def mark_path(initial_area, start_position):
    position = list(start_position)
    area = [list(row) for row in initial_area]

    prev_positions = []
    while position not in prev_positions:
        prev_positions.append(list(position))

        next_position = list(position)
        match position[2]:
            case 0: # UP
                next_position[1] = next_position[1] - 1
            case 1: # RIGHT
                next_position[0] = next_position[0] + 1
            case 2: # DOWN
                next_position[1] = next_position[1] + 1
            case 3: # LEFT
                next_position[0] = next_position[0] - 1

        if not out_of_area(next_position, area) and on_obstacle(next_position, area):
            position[2] = (position[2] + 1) % 4
        else:
            area[position[1]][position[0]] = 'X'
            if out_of_area(next_position, area):
                break
            position = next_position
    
    return area
```

```python
# Part 1

area = read_area(real_input)

start_position = None
for y in range(len(area)):
    if '^' in area[y]:
        start_position = [area[y].index('^'), y, UP]

area = mark_path(area, start_position)

count_x = 0
for row in area:
    print(row)
    count_x += len([x for x in row if x == 'X'])
count_x
```

```python
def assign_symbol(area, position):
    prev_symbol = area[position[1]][position[0]]

    if position[2] == 0:
        next_symbol = '^'
    if position[2] == 1:
        next_symbol = '>'
    if position[2] == 2:
        next_symbol = 'v'
    if position[2] == 3:
        next_symbol = '<'
    
    if (prev_symbol in ['^', 'v'] and next_symbol in ['>', '<']) or (prev_symbol in ['>', '<'] and next_symbol in ['^', 'v']):
        next_symbol = '+'
    elif (prev_symbol == '<' and next_symbol == '>') or (prev_symbol == '>' and next_symbol == '<'):
        next_symbol = '-'
    elif (prev_symbol == '^' and next_symbol == 'v') or (prev_symbol == 'v' and next_symbol == '^'):
        next_symbol = '|'
        
    area[position[1]][position[0]] = next_symbol
    
def check_loop(start_position, area):
    position = list(start_position)
    prev_positions = []
    while position not in prev_positions:
        prev_positions.append(list(position))
        
        next_position = list(position)
        match position[2]:
            case 0: # UP
                next_position[1] = next_position[1] - 1
            case 1: # RIGHT
                next_position[0] = next_position[0] + 1
            case 2: # DOWN
                next_position[1] = next_position[1] + 1
            case 3: # LEFT
                next_position[0] = next_position[0] - 1
        
        if out_of_area(next_position, area):
            return False
        
        match next_position[2]:
            case 0: # UP
                if area[next_position[1]][next_position[0]] == '^':
                    return True
            case 1: # RIGHT
                if area[next_position[1]][next_position[0]] == '>':
                    return True
            case 2: # DOWN
                if area[next_position[1]][next_position[0]] == 'v':
                    return True
            case 3: # LEFT
                if area[next_position[1]][next_position[0]] == '<':
                    return True
        
        if on_obstacle(next_position, area):
            next_position = list(position)
            next_position[2] = (next_position[2] + 1) % 4
        
        position = next_position
    else:
        return True

def mark_smart_path(initial_area, start_position):
    position = list(start_position)
    area = [list(row) for row in initial_area]

    prev_positions = []
    possible_obstacles = []
    while position not in prev_positions:
        prev_positions.append(list(position))

        next_position = list(position)
        match position[2]:
            case 0: # UP
                next_position[1] = next_position[1] - 1
            case 1: # RIGHT
                next_position[0] = next_position[0] + 1
            case 2: # DOWN
                next_position[1] = next_position[1] + 1
            case 3: # LEFT
                next_position[0] = next_position[0] - 1

        if not out_of_area(next_position, area) and on_obstacle(next_position, area):
            assign_symbol(area, position)
            position[2] = (position[2] + 1) % 4
        else:
            assign_symbol(area, position)
            if out_of_area(next_position, area):
                break
            position = next_position

        if check_loop([prev_positions[-1][0], prev_positions[-1][1], (prev_positions[-1][2]+ 1) % 4], area):
            possible_obstacles.append(tuple(position[:2]))

    return area, possible_obstacles
```

```python
# Part 2
initial_area = read_area(real_input)

start_position = None
for y in range(len(initial_area)):
    if '^' in initial_area[y]:
        start_position = [initial_area[y].index('^'), y, UP]

area, possible_obstacles = mark_smart_path(initial_area, start_position)

print(len(possible_obstacles))
for x,y in possible_obstacles:
    area[y][x] = 'O'

for row in area:
    print(row)
```
